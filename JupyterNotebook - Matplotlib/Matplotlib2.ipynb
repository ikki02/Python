```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
plt.style.use('ggplot')
```

```python
# 円弧: .Arc()
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

c1 = mpatches.Arc(xy=(0.4, 0.5), width=0.4, height=0.4, angle=30, theta1=0, theta2=230, linewidth=10, color='lightblue')
c2 = mpatches.Arc(xy=(0.6, 0.5), width=0.4, height=0.4, angle=30, theta1=0, theta2=230, linewidth=10, color='pink')

ax.add_patch(c1)
ax.add_patch(c2)
plt.show()
```

```python
# 矢印: .Arrow()
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

a1 = mpatches.Arrow(x=0.1, y=0.4, dx=0.6, dy=0.5, width=0.5, color='lightblue')
a2 = mpatches.Arrow(x=0.3, y=0.1, dx=0.6, dy=0.5, width=0.5, color='pink')
ax.add_patch(a1)
ax.add_patch(a2)
plt.show()
```

```python
# より詳細な書式設定の矢印: .FancyArrow()
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

a1 = mpatches.FancyArrow(x=0.2, y=0.3, dx=0.4, dy=0.3, width=0.1, head_width=0.3, head_length=0.2, shape='right', color='lightblue')
a2 = mpatches.FancyArrow(x=0.2, y=0.3, dx=0.4, dy=0.3, width=0.1, head_width=0.3, head_length=0.2, shape='left', color='pink')
ax.add_patch(a1)
ax.add_patch(a2)
plt.show()
```

```python
# 円形: .Circle()
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

c1 = mpatches.Circle(xy=(0.5, 0.5), radius=0.4, color='lightblue')
c2 = mpatches.Circle(xy=(0.5, 0.5), radius=0.2, color='white')
ax.add_patch(c1)
ax.add_patch(c2)
plt.show()
```

```python
# 正多角形: .CirclePolygon()
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

cp1 = mpatches.CirclePolygon(xy=(0.5, 0.5), radius=0.4, resolution=6, color='lightblue')
cp2 = mpatches.CirclePolygon(xy=(0.5, 0.5), radius=0.2, resolution=8, color='white')
ax.add_patch(cp1)
ax.add_patch(cp2)
plt.show()
```

```python
# 楕円形: .Ellipse()
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

e1 = mpatches.Ellipse(xy=(0.5, 0.5), width=0.8, height=0.5, angle=30, color='lightblue')
e2 = mpatches.Ellipse(xy=(0.5, 0.5), width=0.5, height=0.2, angle=30, color='white')
ax.add_patch(e1)
ax.add_patch(e2)
plt.show()
```

```python
# 扇形: .Wedge()
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

w1 = mpatches.Wedge(center=(0.3, 0.55), r=0.25, theta1=30, theta2=-30, color='pink')
w2 = mpatches.Wedge(center=(0.7, 0.4), r=0.25, theta1=210, theta2=150, color='lightblue')
ax.add_patch(w1)
ax.add_patch(w2)
plt.show()
```

```python
# 長方形: .Rectangle()
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

r1 = mpatches.Rectangle(xy=(0.1, 0.4), width=0.5, height=0.5, angle=0, color='lightblue')
r2 = mpatches.Rectangle(xy=(0.4, 0.1), width=0.5, height=0.5, angle=0, color='pink')
ax.add_patch(r1)
ax.add_patch(r2)
plt.show()
```

```python
# 多角形: .Polygon()
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

p1 = mpatches.Polygon(xy=[(0, 0.2), (0.3, 0.8), (0.6, 0.2)], facecolor='lightblue')
p2 = mpatches.Polygon(xy=[(0.4, 0.8), (0.7, 0.2), (1.0, 0.8)], facecolor='pink')
ax.add_patch(p1)
ax.add_patch(p2)
plt.show()
```

```python
# フィギュアオブジェクトの書式設定
# 下記以外にsubplotparsという引数があり、左右上下のサブプロット間の余白を設定できる。
# matplotlibrcファイルのデフォルト値も要参考
fig = plt.figure(
                 figsize=(3,3), #サイズ デフォルト値は（6.4, 4.8）
                 dpi=100, #解像度
                 facecolor='grey', #塗りつぶし色
                 edgecolor='k', #枠線の色
                 linewidth=10, #枠線の太さ
                 frameon=True, #枠線を表示する
                 tight_layout=False) #False:subplotparsの設定を使用, True:デフォルト設定
ax1= fig.add_subplot(111)
plt.show()
```

```python
# 色
# 引数として、color, facecolor, edgecolor, alphaがある。alphaは透明度。0は透明。1に近いほど不透明。
x = np.arange(0, 6, 0.1)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure()
ax = fig.add_subplot(111)

ax.scatter(x, y, label='sin', color='lightblue', edgecolor='blue')
ax.scatter(x, z, label='cos', color='lightblue', edgecolor='blue', alpha=0.3)
ax.legend(framealpha=1, loc='lower right')
plt.show()
```

### 指定可能な色の形式例
|名前|例|
|:-:|:-:|
|HTMLカラーネーム|'red','blue','green','DarkOliveGreen'...|
|X11/CSS4カラーネーム|'Red','Blue','Green','DarkSeaGreen'...|
|Matplotlibカラーストリング|'b','g','r','c','m','y'...|
|16進数から―コード|'#4682b4','#ffd700'...|
|正規表現カラー|'C0','C1','C2','C3'...|
他にもいっぱいあるが、多分よく見るのは上のどれかだろう

```python
# カラーマップ：色の組み合わせを定義したもの。
# 指定の仕方として、cm.メソッド()かcm.get_cmap()()の使い方がある。
# いろいろあるが、詳細はググるのがよさそう。
import matplotlib.cm as cm
fig = plt.figure()
ax = fig.add_subplot(111)

x = np.arange(20)
y = np.arange(20)

ax.scatter(x, y+5, color=cm.spring(0.5), label='spring(0.5)')
ax.scatter(x, y, color=cm.get_cmap('winter')(0.5), label='winter(0.5)')

ax.legend()
plt.show()
```

```python
# 段階的に色を変化させる場合
# cmap引数にカラーマップを指定する。
# 下記の場合、引数cに与えたxの値に応じて散布図の要素の色を自動で割り当てる
fig = plt.figure(figsize=(4.5, 2))
ax = fig.add_subplot(111)

x = np.arange(20)
y = np.ones(20)

ax.scatter(x, y, c=x, cmap=cm.Set1)
plt.show()
```

```python
# 段階的に色を変化させる場合（個別に設定する書き方）
fig = plt.figure(figsize=(4.5, 2))
ax = fig.add_subplot(111)

x = np.arange(20)
y = np.ones(20)

for i in range(20):
    color = cm.Set1(i / 20 * 0.5)
    ax.scatter(x[i], y[i], c=color)
plt.show()
```

### 線の書式設定
|オプション|説明|
|:-:|:-:|
|color|線の色|
|linewidth|線の太さ|
|linestyle|線の種類。'-', '--', '-.', ':', 'None'|
|alpha|透明度（1が不透明、0が透明）|
|solid_capstyle|線の先端の形状。butt, round, projectingがある。|
|solid_joinstyle|線の結合点の形状。miter, round, bevel|
|dash_capstyle|破線・点線の先端の形状|
|dash_joinstyle|破線・点線の結合点の形状|

```python
# 線の色と太さを設定する。
x = np.array([0,1])
y = [0,1]

fig = plt.figure(figsize=(9,3))
ax = fig.add_subplot(111, xlim=(-0.2, 4))

ax.plot(x, y, linewidth=1, color='b', label='linewidth=1(alpha=1.0)')
ax.plot(x + 0.1, y, linewidth=1, color='b', alpha=0.5, label='linewidth=1(alpha=0.5)')

ax.plot(x + 0.3, y, linewidth=2, color='b', label='linewidth=2(alpha=1.0)')
ax.plot(x + 0.4, y, linewidth=2, color='b', alpha=0.5, label='linewidth=2(alpha=0.5)')

ax.plot(x + 0.7, y, linewidth=5, color='b', label='linewidth=5(alpha=1.0)')
ax.plot(x + 0.8, y, linewidth=5, color='b', alpha=0.5, label='linewidth=5(alpha=0.5)')

ax.plot(x + 1.1, y, linewidth=10, color='b', label='linewidth=10(alpha=1.0)')
ax.plot(x + 1.3, y, linewidth=10, color='b', alpha=0.5, label='linewidth=10(alpha=0.5)')

ax.legend()
plt.show()
```

```python
# 線の先端の形状
x = [0.1,1]
y = np.array([0,0])

fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot(111, xlim=(0, 1.7), ylim=(-0.2, 1.2))

# フラット
ax.plot(x, y + 1, linewidth=12, solid_capstyle='butt', label='butt')
# 半円
ax.plot(x, y + 0.5, linewidth=12, solid_capstyle='round', label='round')
# 長方形追加
ax.plot(x, y, linewidth=12, solid_capstyle='projecting', label='projecting')

ax.legend(loc=1)
plt.show()
```

```python
# 線の結合点の形状
x = np.array([0, 1, 2, 3])
y = [0, 1.0, 0.2, 1.2]

fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot(111, xlim=(0, 4.5))

# 留め継ぎ
ax.plot(x, y, linewidth=10, solid_joinstyle='miter', label='miter')
# 円形
ax.plot(x + 0.5, y, linewidth=10, solid_joinstyle='round', label='round')
# 面取り
ax.plot(x + 1.0, y, linewidth=10, solid_joinstyle='bevel', label='bevel')

ax.legend()
plt.show()
```

```python
# 線の種類
# linestyleは文字列でも、記号でも、オフセットでも指定できる。
x = np.array([0,1])
y = np.array([1.9, 1.9])

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(111, xlim=(-0.1, 1.7), ylim=(0, 2))

#文字列指定
ax.plot(x, y, linewidth=5, color='b', linestyle='solid', label='solid')
#記号指定
ax.plot(x, y - 0.1, linewidth=5, color='b', alpha=0.5, linestyle='-', label='-')

ax.plot(x, y - 0.4, linewidth=5, color='b', linestyle='dashed', label='dashed')
ax.plot(x, y - 0.6, linewidth=5, color='b', alpha=0.5, linestyle='--', label='--')

ax.plot(x, y - 0.9, linewidth=5, color='b', linestyle='dashdot', label='dashdot')
ax.plot(x, y - 1.0, linewidth=5, color='b', alpha=0.5, linestyle='-.', label='-.')

ax.plot(x, y - 1.3, linewidth=5, color='b', linestyle='dotted', label='dotted')
ax.plot(x, y - 1.4, linewidth=5, color='b', alpha=0.5, linestyle=':', label=':')

#オフセット指定。（offset, (on, off)）。（offset:オフセット、on:描画区間、off:非描画区間）
#5ポイント描画した後、2ポイントの非描画区間を繰り返す。
ax.plot(x, y - 1.7, linewidth=5, color='g', linestyle=(0, (5, 2)), label='(0, (5, 2))')
#オフセットを変える。本来、最初に5ポイント描画されるが、オフセットが4なので、1ポイントしか描画されていない。
ax.plot(x, y - 1.8, linewidth=5, color='g', alpha=0.5, linestyle=(4, (5, 2)), label='(4, (5, 2))')

ax.legend()
plt.show()
```

```python
# 点線破線の先端の形状
x = [0.1, 1]
y = np.array([0, 0])

fig = plt.figure(figsize=(6, 3))
ax = fig.add_subplot(111, xlim=(0, 1.7), ylim=(-0.2, 1.2))

ax.plot(x, y + 1, linewidth=12, linestyle='-.', dash_capstyle='butt', label='butt')
ax.plot(x, y + 0.5, linewidth=12, linestyle='-.', dash_capstyle='round', label='round')
ax.plot(x, y, linewidth=12, linestyle='-.', dash_capstyle='projecting', label='projecting')

ax.legend()
plt.show()
```

```python
# 点線破線の結合点の形状
x = np.array([0, 1, 2, 3])
y = np.array([0, 1.0, 0.2, 1.2])

fig = plt.figure(figsize=(6, 3))
ax = fig.add_subplot(111, xlim=(0, 4.5))

ax.plot(x, y, linewidth=10, linestyle=(0, (3, 0.5)), dash_joinstyle='miter', label='miter')
ax.plot(x + 0.5, y, linewidth=10, linestyle=(0, (3, 0.5)), dash_joinstyle='round', label='round')
ax.plot(x + 1.0, y, linewidth=10, linestyle=(0, (3, 0.5)), dash_joinstyle='bevel', label='bevel')

ax.legend()
plt.show()
```

### テキストボックス
- フォントの書式：matplotlib.font_manager.FontProperties

|オプション|説明|
|:-:|:-:|
|family|フォントファミリーの種類|
|style|フォントスタイル。'normal', 'italic', 'oblique'|
|weight|フォント太さ。0～1000。'ultralight', 'light', 'normal', 'bold'...|
|size|フォントサイズ。'small','medium','large'...|
|alpha|透明度（1が不透明、0が透明）|
|color|フォント色|

- テキストボックスの書式：matplotlib.text.Text

|オプション|説明|
|:-:|:-:|
|bbox|テキストボックスの書式設定|
|verticalalignment|垂直方向の配置。'center', 'top', 'bottom', 'baseline'|
|horizontalalignment|水平方向の配置。'center', 'right', 'left'|
|multialignment|複数行の配置。'center','right','left'|
|linespacing|行間（フォントサイズに対する倍率）|
|fontproperties|フォント書式|
|rotation|回転|
|rotation_mode|回転方法|
|backgroundcolor|背景色|


```python
# テキストを描画する
fig = plt.figure(figsize=(3, 3))
ax = fig.add_subplot(111)

# テキストの文字列左下角の座標を与える
ax.text(2, 4, 'Text', size=40)

# 目盛を入力する。
ax.set_xticks([i * 2 for i in range(6)])
ax.set_yticks([i * 2 for i in range(6)])

# 目盛線描画
ax.grid(linestyle='-')
plt.show()
```

```python
# フォントの書式を設定する
fig = plt.figure(figsize=(3, 3))
ax = fig.add_subplot(111)

# フォント設定をしてテキストを描画
ax.text(0.2, 0.6, 'Text', alpha=0.5, color='Red', size=40)

# フォント設定をしてテキストを描画（辞書使用）
fps = {
    'family': 'fantasy',
    'style': 'italic',
    'weight': 'heavy',
    'size': 40,
    'color': 'Green'
    }

# 辞書ではcolorはGreenがデフォルトとして与えられるが、引数で直接上書きできる。
ax.text(0.2, 0.2, 'Text', alpha=0.5, color='Red', fontdict=fps)

ticks = [0.2 * i for i in range(6)]
ax.set_xticks(ticks)
ax.set_yticks(ticks)

# 目盛線描画
ax.grid(linestyle='-')
plt.show()
```

```python
# テキストボックスの書式設定
fig = plt.figure(figsize=(3, 3))
ax = fig.add_subplot(111)

# テキストボックスの書式辞書
boxprops = {
    'facecolor': 'pink',
    'edgecolor': 'red',
    'alpha': 0.5,
    'boxstyle': 'Sawtooth',
    'linewidth': 2
    }

# テキストボックスの描画
ax.text(0.07, 0.5, 'テキストボックス\nの書式設定による\n設定', bbox=boxprops, size=18)

ax.text(0.07, 0.2, '塗りつぶし色の設定', bbox={'alpha': 0.2}, backgroundcolor='Blue', size=16)

ticks = [0.2 * i for i in range(6)]
ax.set_xticks(ticks)
ax.set_yticks(ticks)

# 目盛線描画
ax.grid(linestyle='-')
plt.show()
```

```python
# テキストボックスの水平方向の配置を設定する。
fig = plt.figure(figsize=(3, 3))
ax = fig.add_subplot(111)

x, y = [0.5, 0.5], [0, 4]
ax.plot(x, y)

# 座標にテキストボックスの左下角、中央下、右下角をそれぞれ合わせる。
ax.text(0.5, 3, 'left', horizontalalignment='left', size=20)
ax.text(0.5, 2, 'center', horizontalalignment='center', size=20)
ax.text(0.5, 1, 'right', horizontalalignment='right', size=20)

ax.set_xticks([0.5 * i for i in range(3)])
ax.set_yticks([i for i in range(5)])

plt.show()
```

```python
# テキストボックスの垂直方向の配置を設定する。
fig = plt.figure(figsize=(8, 3))
ax = fig.add_subplot(111)

x, y = [0, 9], [0.5, 0.5]
ax.plot(x, y)

# 座標にテキストボックスの左下角(bottom)、文字列の左下角(baseline)、
# テキストボックス左端・Y軸方向の中央(center)、右上角(top)をそれぞれ合わせる。
ax.text(0, 0.5, 'bottom', verticalalignment='bottom', size=20)
ax.text(2.5, 0.5, 'baseline', verticalalignment='baseline', size=20)
ax.text(5.5, 0.5, 'center', verticalalignment='center', size=20)
ax.text(8, 0.5, 'top', verticalalignment='top', size=20)

ax.set_xticks([i * 3 for i in range(4)])
ax.set_yticks([i * 0.5 for i in range(3)])

plt.show()
```

```python
# テキストボックス内の文字列の水平方向の配置を設定する。
fig = plt.figure(figsize=(4, 7))
ax = fig.add_subplot(111, xlim=[0, 5], ylim=[0, 14])

# 文字列を左揃え、中央揃え、右揃えにそれぞれ合わせる。
ax.text(1, 10, '左揃え\nLeft\nalignment', multialignment='left', size=20)
ax.text(1, 6, '中央揃え\nCenter\nalignment', multialignment='center', size=20)
ax.text(1, 2, '右揃え\nRight\nalignment', multialignment='right', size=20)

ax.grid(linestyle='-')
plt.show()
```

```python
# テキストボックス内の文字列の行間を設定する
# linespacingはフォントサイズに対する倍率で設定する。
fig = plt.figure(figsize=(3.5, 5))
ax = fig.add_subplot(111, xlim=[0, 4], ylim=[0, 7])

ax.text(1, 5, '行間 1\nspace 1', linespacing=1, size=20)
ax.text(1, 3, '行間 1.5\nspace 1.5', linespacing=1.5, size=20)
ax.text(1, 1, '行間 2\nspace 2', linespacing=2, size=20)

ax.grid(linestyle='-')
plt.show()
```

```python
# テキストボックスの回転（回転の向きは左回転である。）
# ax1: 回転しない場合。rotation=0
# ax2: 90度回転する場合。rotation=90
# ax3: テキストボックス左下角を軸に45度回転する場合。rotation=45, rotation_mode='anchor'
# ax4: 図形の中心を軸に45度回転する場合。rotation=45, rotation_mode='None'
# ax3とax4の違いに注意。
fig = plt.figure(figsize=(8, 8))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

axes = [ax1, ax2, ax3, ax4]
x, y = [0.5, 0.5], [0, 1]

for ax in axes:
    ax.plot(x, y, color='SteelBlue')
    ax.plot(y, x, color='SteelBlue')

for ax in axes:
    ax.set_xticks([0.5 * i for i in range(3)])
    ax.set_yticks([0.5 * i for i in range(3)])

ax1.text(0, 0.9, '1)', size=16)
ax2.text(0, 0.9, '2)', size=16)
ax3.text(0, 0.9, '3)', size=16)
ax4.text(0, 0.9, '4)', size=16)

ax1.text(0.5, 0.5, 'horizontal', rotation='horizontal', size=12) #rotation=0と同義
ax2.text(0.5, 0.5, 'vertical', rotation='vertical', size=12) #rotation=90と同義
ax3.text(0.5, 0.5, 'rotation=45\nwith anchor\nsetting', rotation=45, rotation_mode='anchor', size=12)

# ax4の回転前の書式設定
boxprops={
    'facecolor': 'Pink',
    'edgecolor': 'Pink',
    'alpha': 0.5
    }
ax4.text(0.5, 0.5, 'rotation=45\nwithout anchor\nsetting', rotation=0, bbox=boxprops, color='white', size=12)

# ax4の回転後の書式設定
boxprops={
    'facecolor': 'None',
    'edgecolor': 'Pink',
    }
ax4.text(0.5, 0.5, 'rotation=45\nwithout anchor\nsetting', rotation=45, bbox=boxprops, size=12)
    
plt.show()
```

### グラフの軸と目盛の設定

```python
# 軸の一括範囲設定：.set_xlim()
# set_xlim()を使わないと、自動で範囲を判定してくれるが、比較のためには揃えた方がよい。
x1 = np.arange(0, 15, 0.2)
x2 = np.arange(0, 20, 0.2)

y1 = np.sin(x1)
y2 = np.sin(x2) * 2
y3 = np.cos(x2) * 2
y4 = np.cos(x1)

fig = plt.figure(figsize=(3, 3))

ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

ax1.plot(x1, y1, label='y1')
ax2.plot(x2, y2, label='y2')
ax3.plot(x2, y3, label='y3')
ax4.plot(x1, y4, label='y4')

axes = [ax1, ax2, ax3, ax4]
for ax in axes:
    ax.set_xlim(0, 20)
    ax.set_ylim(-2, 2)

fig.subplots_adjust(wspace=0.35, hspace=0.35, bottom=0.2)
plt.show()
```

```python
# 軸の設定をそろえる。subplots()の場合。
fig, axes = plt.subplots(2, 2, figsize=(3, 3), sharex=True, sharey=True)

axes[0, 0].plot(x1, y1)
axes[0, 1].plot(x2, y2)
axes[1, 0].plot(x2, y3)
axes[1, 1].plot(x1, y4)
plt.show()
```

```python
# 対数軸を設定する。.set_yscale('log')
x1 = np.linspace(0, 100, 100)
x2 = np.linspace(0, 1000000, 100)
y1 = np.linspace(0, 1000000, 100)
y2 = np.linspace(0, 100, 100)

fig = plt.figure(figsize=(8, 2))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.plot(x1, y1)
ax2.plot(x2, y2)

ax1.set_yscale('log')
ax2.set_xscale('log')

ax1.set_title('Y軸が対数軸')
ax2.set_title('X軸が対数軸')
plt.show()
```

```python
# 複数軸を設定する。.twinx()
# twinx()で2つのY軸を共有するX軸を作る。
x1 = np.linspace(0, 10, 100)
x2 = np.linspace(0, 100, 100)
y1 = np.linspace(0, 100, 100)
y2 = np.linspace(0, 10, 100)

fig = plt.figure(figsize=(9, 3))
ax1_1 = fig.add_subplot(121)
ax2_1 = fig.add_subplot(122)

ax1_1.plot(x1, y1)
ax1_2 = ax1_1.twinx()
ax1_2.plot(x1, y2)

ax2_1.plot(x1, y1)
ax2_2 = ax2_1.twiny()
ax2_2.plot(x2, y1)

ax1_1.set_title('2つのY軸', y=0.8)
ax2_1.set_title('2つのX軸', y=0.8)
plt.show()
```

```python
# 軸ラベルを設定する。.set_xlabel()
x = np.arange(0, 15, 0.2)
y = np.sin(x)

fig = plt.figure(figsize=(4, 2))
ax = fig.add_subplot(111)

ax.plot(x, y)

ax.set_xlim([0, 15])
ax.set_ylim([-1, 1])

ax.set_xlabel('x', fontsize=20, color='red')
ax.set_ylabel('y', {'fontsize': 20, 'color': 'green'})

fig.subplots_adjust(bottom=0.3, left=0.2)
plt.show()
```

```python
# 主目盛と主目盛ラベルの設定
# set_xticksには任意の要素数のリストを渡せる。数値、もしくは真偽値
# set_xtickslabelsには主目盛と同じ要素数のリストを渡せる。数値、テキストもしくは真偽値
# set_xtickslabelsに空のリストを渡すと、ラベルを非表示にできる。
fig = plt.figure(figsize=(4, 2))
ax = fig.add_subplot(111)

ax.plot(x, y)

ax.set_xlim([0, 15])
ax.set_ylim([-1, 1])

ax.set_xticks([0, 5, 10, 15])
ax.set_yticks([-1, 0, 1])

ax.set_xticklabels(['zero', 'five', 10, 15])
ax.set_yticklabels([])

fig.subplots_adjust(bottom=0.3, left=0.2)
plt.show()
```

|zorder|描画順|

|right|右軸目盛（'on','off'）|

```python
# 詳細設定
fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot(111)

# 補助目盛の描画
ax.minorticks_on()

# X軸の補助目盛を設定
ax.tick_params(axis='x', which='minor', bottom='off')

# X軸の主目盛を設定
ax.tick_params(axis='x', which='major', direction='inout', length=10, labelcolor='green')

# Y軸の補助目盛を設定
ax.tick_params(axis='y', which='minor', direction='out', length=5, color='red')

# Y軸の主目盛を設定
ax.tick_params(axis='y', which='major', direction='inout', length=10, labelcolor='blue')

plt.show()
```

### 凡例とタイトル

↓figureオブジェクトのlegend()引数

|引数|意味|
|:-:|:-:|
|loc|凡例のアンカの位置。下記コードまたは文字列で指定できる。|
|bbox_to_anchor|アンカの座標。サブプロットの左下角は（0,0）、右上角は（1,1）|

|コード|文字列|
|:-:|:-:|
|0|best|
|1|upper right|
|2|upper left|
|3|lower left|
|4|lower right|
|5|right|
|6|center left|
|7|center right|
|8|lower center|
|9|upper center|
|10|center|

```python
# plot()のlabel引数 および figureオブジェクトのlegend()
# デフォルトでは凡例はサブプロットの内側に配置される。
x = np.arange(0, 20, 0.2)
y1 = np.sin(x)
y2 = np.cos(x)

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')
ax.legend()
plt.show()
```

```python
# 凡例を配置する。
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')

ax.legend(loc='upper left', bbox_to_anchor=(1, 1)) #bbox_to_anchorがlocより優先される。そのため、右上角に凡例がくる。
fig.subplots_adjust(left=0.2, bottom=0.2, right=0.75, top=0.8) #サブプロットの位置の調整。ファイル出力時これが効く。
plt.show()
```

```python
# 凡例の位置ずれを調整する。
# borderaxespadはサブプロットの枠線と凡例の枠線との境界領域の幅を設定する。デフォルトは0.5
# 上図と比べて凡例の位置が少し変わったことに気づこう。
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')

ax.legend(loc='upper left', bbox_to_anchor=(1, 1), borderaxespad=0) 
fig.subplots_adjust(left=0.2, bottom=0.2, right=0.75, top=0.8) #サブプロットの位置の調整。ファイル出力時これが効く。
plt.show()
```

```python
# サブプロットのタイトル
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')

ax.set_title('sin and cos')
plt.show()
```

```python
# フィギュアのタイトルを設定する。
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')

ax.set_title('sin and cos')
fig.suptitle('三角関数')
plt.show()
```

```python
# 凡例とタイトルの書式設定を設定する
# フォントとテキストボックスの書式設定と同様の設定ができる。
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')

# フォントの書式辞書
fdict = {'family': 'serif',
         'style': 'italic',
         'weight': 'bold',
         'size': 12}

# テキストボックスの書式辞書
bdict = {'facecolor': 'k',
         'edgecolor': 'None',
         'alpha': 0.5,
         'boxstyle': 'rarrow',
         'linewidth': 2}

ax.legend(loc='upper left', bbox_to_anchor=(1.02, 1), borderaxespad=0, prop=fdict)
ax.set_title('sin and cos', fontdict=fdict, bbox=bdict, color='white')
fig.suptitle('三角関数', y=0.95, size=16, color='grey', fontweight='bold')
fig.subplots_adjust(left=0.2, bottom=0.2, right=0.75, top=0.8)

plt.show()
```
